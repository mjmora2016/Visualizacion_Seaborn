## Data Visualization with Seaborn

https://www.analyticsvidhya.com/blog/2019/09/comprehensive-data-visualization-guide-seaborn-python/

Why Seaborn? Seaborn is a visualization library for python that gives the user the power of creating great visualizations in an easy way and works well with Dataframes. It's built in a high layer on Matplotlib so hard to perform visualizations with matplotlib could be easy to create with Seaborn.

__1.- Imports__

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from scipy import stats
```

```python
%matplotlib inline
```

```python
df = pd.read_csv("data/StudentsPerformance.csv", header='infer')
```

```python
df[0:5]
```

```python
df.info(memory_usage=True)
```

__2.- Looking for relationships:__

__Scatterplot:__

```python
sns.relplot( x='reading score', y='writing score', hue='gender', data = df)
```

__3.- Plotting Categorical Data:__

Categorical vs Numerical

__Jitter Plot__

```python
sns.catplot(x='gender', y='math score', jitter = True, data = df)
```

__Hue Plot__

```python
sns.catplot(x="gender", y="math score", hue = "race/ethnicity", data=df)
```

To avoid overlaping set parameter **kind** to  _swarm_

```python
sns.catplot(x="gender", y="math score", hue = "race/ethnicity", kind = "swarm", data=df)
```

__Boxplots__

To get the quantile visualization we use Boxplots setting the **kind** parameter to _box_...

```python
sns.catplot(x="gender", y="math score", hue = "race/ethnicity", kind = "box", data=df)
```

... set it to _violin_ to get a Violin representation

```python
sns.catplot(x="gender", y="math score", hue = "race/ethnicity", kind = "violin", data=df)
```

If we set the **kind** parameter to _bar_ then the bar will show the mean value:

```python
sns.catplot(x="gender", y="math score", hue = "race/ethnicity", kind = "bar", data=df)
```

**Pointplot**

The value **point** in _kind_ parameter shows the points plot including the mean value and the representation of the confidence interval:

```python
sns.catplot(x="math score", y="gender", hue = 'test preparation course', kind = "point", data=df)
```

**4.- Plotting Distributions**

**Histogram**

To plot the density distribution (kde parameter default value is True):

```python
sns.distplot(df['math score'])
```

```python
sns.distplot(df['math score'], kde=False)
```

**Bivariate Distributions**

To plot the realtion between to variables and also plot the distribution:

```python
sns.jointplot(x="reading score", y="writing score", data=df);
```

With the __kind__ parameter set to _hex_ value we can plot, the relatioship between two variables, their distributions and and hexagonal colormap that identify  the number of observations in a hexagon by its color.

```python
sns.jointplot(x="reading score", y="writing score", kind = "hex", data=df);
```
